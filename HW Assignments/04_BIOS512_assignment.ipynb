```r
library(tidyverse)

library(readr) 

windturbine = read_csv("https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-10-27/wind-turbine.csv", 
                       col_types = cols())
print(head(windturbine, 10))
```

Each row in the table represents a single turbine.

Make a bar chart ðŸ“Š **sorted by frequency** of turbine count per province territory. (Hint: `fct_infreq` will sort your factor levels by frequency for subsequent plotting)

```r
p =  ggplot(windturbine, aes(x = fct_infreq(province_territory))) +
  geom_bar() +
  labs(title = "Turbine Count per Province/Territory",
       x = "Province/Territory",
       y = "Turbine Count")+ theme(axis.text.x = element_text(angle = 90, hjust = 1))

p
```

Make a bar chart **sorted by frequency** of turbine count per manufacturer. **Group the low frequency manufacturers into an `Other` category**.

**Bonus:** Can you force the `Other` category to be the last bar? 

```r
windturbine$manufacturer <- fct_lump(windturbine$manufacturer, n = 5, other_level = "Other")

p <- ggplot(windturbine, aes(x = fct_infreq(manufacturer))) +
  geom_bar() +
  labs(title = "Turbine Count per Manufacturer",
       x = "Manufacturer",
       y = "Turbine Count")

print(p)
```
